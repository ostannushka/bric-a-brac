С датами немного сложно: они пригодятся для генерирования фичей, но модели машинного обучения с форматом datetime не работают. Кроме того важна не сама дата регистрации пользователя, а срок его пользования услугами провайдера на момент анализа (1 февраля 2020). Поэтому столбец BeginDate будет переведен в формат datetime, и будет сформирован новый столбец refused, который собственно и будет целевым признаком. Помимо этого, необходимо вычислить время пользования услугами на момент отказа или на текущий момент (1.02.2020), если пользователь не расторг контракт.

```python
#функция для заполнения столбца refused
def is_refused(row):
    if row['EndDate'] == 'No':
        return 0
    else:
        return 1

contract['refused'] = contract.apply(is_refused, axis=1)
```

```python
contract['BeginDate'] = pd.to_datetime(contract['BeginDate'], format='%Y-%m-%d')
print(type(contract['BeginDate'][1]))
# теперь заменим столбец EndDate на время пользования
# если оставить этот столбец в текущем виде, в него будет "утекать" целевой признак
def how_many(row):
    if row['EndDate'] != 'No':
        row['EndDate'] = pd.to_datetime(row['EndDate'], format='%Y-%m-%d')
        return row['EndDate'] - row['BeginDate']
    else:
        return pd.to_datetime('2020-02-01', format='%Y-%m-%d') - row['BeginDate']

contract['tenure'] = contract.apply(how_many, axis=1)
# на данном этапе можно было бы удалить столбец EndDate, но он понадобится для анализа данных, 
# поэтому столбцы с датами будут удалены непосредственно перед обучением моделей
```

```python
# данные в столбце Tenure относятся к типу timedelta64[ns], это плохо, т.к. из них не построить диаграмму
contract['tenure'] = contract['tenure'].dt.days
```

#### Анализ категориальных переменных

```python
def draw_pie(df, col, title):
    for_pie = df.groupby(col).size().reset_index(name='counts')

    # Draw Plot
    fig, ax = plt.subplots(figsize=(8, 5), subplot_kw=dict(aspect="equal"), dpi= 80)

    data = for_pie['counts']
    categories = for_pie[col]

    def func(pct, allvals):
        absolute = int(pct/100.*np.sum(allvals))
        return "{:.1f}% ({:d} )".format(pct, absolute)

    wedges, texts, autotexts = ax.pie(data, 
                                  autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"), 
                                  colors=plt.cm.Dark2.colors,
                                 startangle=140)
    # Decoration
    ax.legend(wedges, categories, title=col, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    plt.setp(autotexts, size=10, weight=700)
    plt.title(title)
    plt.show()
```

```python
# Type - тип оплаты
draw_pie(contract, 'Type', 'Все пользователи')
draw_pie(not_refused, 'Type', 'Продолжают пользоваться услугами компании')
draw_pie(refused, 'Type', 'Отказались от услуг')
```

Среди всех пользователей преобладает помесячная оплата, а далее примерно поровну разделились пользователи, котрые оплачивают услуги раз в год и раз в два года. Среди продолжающих пользоваться услугами больше пользователей с длительным периодом оплаты, а среди отказавшихся, наоборот, больше пользователей с помесячной оплатой. Это важно, т.к. возможно стимулирование пользователей к переходу на длительный период оплаты с помощью рекламных кампаний и скидок поможет сохранить эту аудиторию.

```python
# PaperlessBilling - электронный чек
draw_pie(contract, 'PaperlessBilling', 'Все пользователи')
draw_pie(not_refused, 'PaperlessBilling', 'Продолжают пользоваться услугами компании')
draw_pie(refused, 'PaperlessBilling', 'Отказались от услуг')
```

Большинство пользователей предпочитают электронный чек. Среди пользователей, отказавшихся от услуг компании эта разница более заметна.

#### Выдвижение и проверка гипотез

По данным графического анализа формируются несколько гипотез о влиянии признаков на целевой. Некоторые факторы выглядят случайными, некоторые могут действительно влиять на развитие целевого события. 
1. Время пользования услугами компании в группе refused меньше, чем у остальных пользователей
2. Ежемесячные расходы на связь в группе refused в среднем больше, чем у остальных
3. Пользователей, которые оплачивают связь наперёд (за год или за два), в группе refused значительно меньше
4. Количество дополнительных услуг из раздела internet в среднем меньше в группе refused, чем у остальных пользователей

```python
# Время пользования услугами компании в группе refused меньше, чем у остальных пользователей
print(refused['tenure'].mean())
print(not_refused['tenure'].mean())

# нулевая гипотеза: средние генеральных совокупностей равны
# альтернативная(односторонняя): среднее в группе refused значимо меньше

alpha = 0.01 # уровень статистической значимости

results = st.ttest_ind(refused['tenure'], not_refused['tenure'])
print('p-значение:', results.pvalue / 2) # гипотеза односторонняя, поэтому р пополам

if results.pvalue/2 < alpha:
    # сравним получившееся p-значение с заданным уровнем статистической значимости 
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Да, время пользования услугами компании в группе refused меньше, чем у остальных пользователей

```python
# Ежемесячные расходы на связь в группе refused в среднем больше, чем у остальных
print(refused['MonthlyCharges'].mean())
print(not_refused['MonthlyCharges'].mean())

# нулевая гипотеза: средние генеральных совокупностей равны
# альтернативная(односторонняя): среднее в группе refused значимо больше

alpha = 0.01 # уровень статистической значимости

results = st.ttest_ind(refused['MonthlyCharges'], not_refused['MonthlyCharges'])
print('p-значение:', results.pvalue / 2) # гипотеза односторонняя, поэтому р пополам

if results.pvalue/2 < alpha:
    # сравним получившееся p-значение с заданным уровнем статистической значимости
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

И здесь разница ощутима

```python
# Пользователей, которые оплачивают связь наперёд (за год или за два), в группе refused значительно меньше
# здесь переменные категориальные, поэтому придется считать хи-квадрат
# но сначала уменьшим коичество классов с 3 до 2, с бинарными признаками работать всегда лучше
contract['Type'].unique()
def change_type_for_stats(col):
    if col == 'Two year' or col == 'One year':
        return "longtime"
    else:
        return col
type_for_stats = contract[['Type', 'refused']]
type_for_stats['Type'] = type_for_stats['Type'].apply(change_type_for_stats)
contigency= pd.crosstab(type_for_stats['Type'], type_for_stats['refused']) 
contigency # таблица сопряженности
```

```python
# нулевая гипотеза в данном случае звучит: между этими переменными нет связи
# альтернативная - переменные взаимосвязаны
stat, p, dof, expected = st.chi2_contingency(contigency) # вызываем хи-квадрат для сопряженной таблицы
# эта функция возвращает  chi, p, dof(degree of freedom) and expected(Frequencies to expect)
print(p < 0.05)
```

Нулевую гипотезу можно отвергнуть - связь между категориальными переменными (тип оплаты и отказ от услуг) значима

```python
#random forest
model_1 = None
best_result = 0
balance = ['balanced', None]
for bal in balance:
    for est in range(105, 108, 1):
        for d in range(8, 11):
            model = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth = d, class_weight= bal) # обучение модели с заданным количеством деревьев
            model.fit(features_train_ohe, target_train) # обучение модели на тренировочной выборке
            probabilities_valid = model.predict_proba(features_valid_ohe)
            probabilities_one_valid = probabilities_valid[:, 1]
            # посчитаем площадь под ROC кривой
            result = roc_auc_score(target_valid, probabilities_one_valid)
            if result > best_result:
                model_1 = model
                best_result = result
                n_est = est
                depth = d
                weight = bal

print("roc-auc лучшей модели на валидационной выборке:", best_result, "Количество деревьев", n_est, "Глубина дерева", d, "вес классов", weight)
```
